# ðŸ“Œ **Install & Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-darkgrid")
sns.set_context("notebook")
```

# **ðŸ“Œ Load Dataset**

```python
df = pd.read_csv("apple_products.csv")  # Upload via Colab sidebar
df.head()
```

# **ðŸ“Œ Distribution of Sale Prices**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sale Price'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Sale Prices")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

# **ðŸ“Œ Top 20 Products by Discount**

```python
top_discounts = df[['Product Name', 'Discount Percentage']].dropna().sort_values(by='Discount Percentage', ascending=False).head(20)
plt.figure(figsize=(10, 8))
sns.barplot(data=top_discounts, x='Discount Percentage', y='Product Name', palette='viridis')
plt.title("Top 20 Products by Discount Percentage")
plt.xlabel("Discount Percentage")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

# **ðŸ“Œ Star Ratings Distribution**

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Star Rating', palette='pastel')
plt.title("Distribution of Star Ratings")
plt.xlabel("Star Rating")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

# **ðŸ“Œ Top 15 Products by Reviews**

```python
top_reviews = df[['Product Name', 'Number Of Reviews']].dropna().sort_values(by='Number Of Reviews', ascending=False).head(15)
plt.figure(figsize=(10, 8))
sns.barplot(data=top_reviews, x='Number Of Reviews', y='Product Name', palette='magma')
plt.title("Top 15 Products by Number of Reviews")
plt.xlabel("Number Of Reviews")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

# **ðŸ“Œ Rating vs Price by RAM**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Star Rating', y='Sale Price', hue='Ram', palette='coolwarm', alpha=0.7)
plt.title("Star Rating vs. Sale Price Colored by RAM")
plt.xlabel("Star Rating")
plt.ylabel("Sale Price")
plt.legend(title="RAM")
plt.tight_layout()
plt.show()
```
